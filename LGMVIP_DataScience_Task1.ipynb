**Image to Pencil Sketch with Python**

Packages used:  **OpenCV**

Installing required libraries

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("pikapika.jpg")
cv2_imshow(img)
```

Converting the image to grayscale

```python
grayimg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(grayimg)
```

Inverting the image: finding negative image by subtracting pixel value from 255

```python
imginverted = 255 - grayimg
cv2_imshow(imginverted)
```

Applying GaussianBlur

```python
blur = cv2.GaussianBlur(imginverted,(21,21),0)
inverted = 255-blur
cv2_imshow(inverted)
```

Dodging the image

```python
def dodgeV2(x,y):
  return cv2.divide(x,255-y,scale=256)
```

```python
final = dodgeV2(grayimg,inverted)
cv2_imshow(final)
```
